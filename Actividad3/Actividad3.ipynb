```python
#Vamos a importar las bibiliotecas de pandas, numpy y matplot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#vamos a leer los dos archivos de Abha para poder trabajar con ellos
Dec0=pd.read_csv('AbhaDecember.txt', skiprows=8, sep='\s+')
Jun0=pd.read_csv('AbhaJune.txt', skiprows=9, sep='\s+')
#Como nos saltamos los títulos de las columnas, las volvemos a nombrar:
Dec0.columns=['PRES','HGHT','TEMP','DWPT','RELH','MIXR','DRCT','SKNT','THTA','THTE','THTV']
Jun0.columns=['PRES','HGHT','TEMP','DWPT','RELH','MIXR','DRCT','SKNT','THTA','THTE','THTV']
```

```python
#No todo el documento es la tabla de datos que queremos usar, así que vamos a droppear la información que hay debajo de 
#las tablas.
#
#Drop sirve para slaltarnos cierto número de filas, no las eliminamos del documento, simplemente las saltamos al momento
#de leerlas.
Dec0=Dec0[:-21]
```

```python
#Queremos leer nuestros datos en la tabla, pero estos se leen como objeto (como el tipo CHARACTER de Fortran).
#Para convertirlos de objeto a flotante se usa la función:
#
#Primero con Diciembre
ConjuntoCol=['PRES','HGHT','TEMP','DWPT','RELH','MIXR','DRCT','SKNT','THTA','THTE','THTV']
Dec0[ConjuntoCol]=Dec0[ConjuntoCol].apply(pd.to_numeric,errors='coerce',axis=1)
```

```python
Jun0.head(100)
```

```python
#Luego con Junio
ConjuntoCol=['PRES','HGHT','TEMP','DWPT','RELH','MIXR','DRCT','SKNT','THTA','THTE','THTV']
Jun0[ConjuntoCol]=Jun0[ConjuntoCol].apply(pd.to_numeric,errors='coerce',axis=1)
```

```python
#Vamos a darle estructura a los datos
#
#Primero Diciembre
Dec=pd.DataFrame(Dec0)
```

```python
#Luego Junio
Jun=pd.DataFrame(Jun0)
```

```python
#Vamos a empezar a gráficar.
#
#La primera es una gráfica de presión como función de la altura
#
#Diciembre
Dec1=Dec[['HGHT','PRES']]
plt.figure(); Dec1.plot(x='PRES'); plt.legend(loc='best')
plt.title('Gráfica de Presión como Función de la Altura')
plt.xlabel('Presión (hPa)')
plt.ylabel('Altura (m)')
plt.xlim([100,800])
plt.grid(True)
plt.show()
```

```python
#Junio
Jun1=Jun[['HGHT','PRES']]
plt.figure(); Jun1.plot(x='PRES'); plt.legend(loc='best')
plt.title('Gráfica de Presión como Función de la Altura')
plt.xlabel('Presión (hPa)')
plt.ylabel('Altura (m)')
plt.xlim([0,800])
plt.grid(True)
plt.show()
```
