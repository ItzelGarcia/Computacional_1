```python
#Para poder graficar x(t), y(t) y z(t) contra el tiempo, necesitamos obtener las ecuaciones.
#Para eso, vamos a utilizar las primeras celdas de "1-Visualización" ya que ahí se obtienen las pocisiones al integrar las derivadas
#usando ODIENT.

#De la animación 1:

%matplotlib inline
import numpy as np, matplotlib.pyplot as plt, matplotlib.font_manager as fm, os
from scipy.integrate import odeint
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
font_family = 'Myriad Pro'
title_font = fm.FontProperties(style='normal', size=20, weight='normal', stretch='normal')
```

```python
save_folder = 'images'
if not os.path.exists(save_folder):
    os.makedirs(save_folder)
```

```python
# define the initial system state (aka x, y, z positions in space)
initial_state = [0.1, 0, 0]

# define the system parameters sigma, rho, and beta
sigma = 10.
rho   = 28.
beta  = 8./3.

# define the time points to solve for, evenly spaced between the start and end times
start_time = 0
end_time = 100
time_points = np.linspace(start_time, end_time, end_time*100)
```

```python
# define the lorenz system
# x, y, and z make up the system state, t is time, and sigma, rho, beta are the system parameters
def lorenz_system(current_state, t):
    
    # positions of x, y, z in space at the current time point
    x, y, z = current_state
    
    # define the 3 ordinary differential equations known as the lorenz equations
    dx_dt = sigma * (y - x)
    dy_dt = x * (rho - z) - y
    dz_dt = x * y - beta * z
    
    # return a list of the equations that describe the system
    return [dx_dt, dy_dt, dz_dt]
```

```python
# use odeint() to solve a system of ordinary differential equations
# the arguments are: 
# 1, a function - computes the derivatives
# 2, a vector of initial system conditions (aka x, y, z positions in space)
# 3, a sequence of time points to solve for
# returns an array of x, y, and z value arrays for each time point, with the initial values in the first row
xyz = odeint(lorenz_system, initial_state, time_points)

# extract the individual arrays of x, y, and z values from the array of arrays
x = xyz[:, 0]
y = xyz[:, 1]
z = xyz[:, 2]
```

```python
from numpy import loadtxt
from pylab import figure, plot, xlabel, xlim, ylim, ylabel, grid, hold, legend, title, savefig
from matplotlib.font_manager import FontProperties
%matplotlib inline


f1 = figure(figsize = (8,4))

#Vamos a graficar el tiempo contra y(t):


plot(time_points, y, 'tab:pink')


#Vamos a graficar el tiempo contra x(t):


plot(time_points, x, 'b-')


#Vamos a graficar el tiempo contra z(t):


plot(time_points, z, 'tab:cyan')
grid()
xlabel('time')
ylabel('$x(t), y(t), z(t)$')
title('Gráfica 1 Bidimensional de las Soluciones x(t), y(t), z(t)')
f1.savefig('2.graficas1.png')
```

```python
#Podemos ver que no hay una clara distinción entre la x(t) y y(y), así que vamos a graficar ambas con un zoom.


#tiempo contra y(t):


plot(time_points, y, 'tab:pink')
grid()
xlabel('time')
ylabel('$y$')
xlim(20,30)


#tiempo contra x(t):


plot(time_points, x, 'b-')
grid()
xlabel('time')
ylabel('$x$')
xlim(20,30)


#tiempo contra z(t):


plot(time_points, z, 'tab:cyan')
grid()
xlabel('time')
ylabel('$z$')
xlim(20,30)
title('Zoom de Gráfica 1 Bidimensional de las Soluciones x(t), y(t), z(t)')
f1.savefig('2.zoom1.png')
```
